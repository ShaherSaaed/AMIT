```python
color_counts = df['color'].value_counts()
common_colors = color_counts[color_counts > 65]
other = color_counts[color_counts <= 65].sum()
colors = pd.concat([common_colors, pd.Series([other], index=['Other'])])
colors
```

```python
explode = (0.1, 0, 0, 0, 0, 0, 0)
plt.figure(figsize=(8,8))
plt.pie(colors.values, labels=colors.index, autopct='%1.1f%%', explode=explode)
plt.show()
```
