```python
import pandas as pd
from sqlalchemy import create_engine, text
from urllib.parse import quote_plus
```

```python
password = quote_plus("MySql@2024!")
engine = create_engine(f"mysql+pymysql://root:{password}@localhost:3306/bookstore_db")
```

```python
with engine.begin() as conn:
    conn.execute(text("""
        INSERT INTO categories (category_name) VALUES
        ('Science Fiction'),
        ('Biography'),
        ('Business');
"""))

    conn.execute(text("""
        INSERT INTO books(title, price, category_id) VALUES
        ('The Martian', 150.00, 1),
        ('Elon Musk Biography', 200.00, 2),
        ('Steve Jobs Biography', 210.00, 2),
        ('Start With Why', 120.00, 3);
"""))
    
    conn.execute(text("""
        INSERT INTO orders (book_id, quantity, order_date) VALUES
        (1, 2, '2024-06-01'),
        (2, 1, '2024-06-02'),
        (3, 3, '2024-06-03');
"""))
```

```python
with engine.begin() as conn:
        result = conn.execute(text("""
        SELECT title, price, order_id
        FROM books B
        JOIN orders O
        ON B.book_id = O.book_id;"""))
        df = pd.DataFrame(result.fetchall(), columns=result.keys())
        display(df)
```

```python
with engine.begin() as conn:
        result = conn.execute(text("""
        SELECT title, price FROM books
        WHERE price > 150.00;"""))
        df = pd.DataFrame(result.fetchall(), columns=result.keys())
        display(df)
```

```python
with engine.begin() as conn:
    result = conn.execute(text("""
    SELECT category_name, COUNT(category_id) AS 'number of books per category'
    FROM categories
    GROUP BY category_name;"""))
    df = pd.DataFrame(result.fetchall(), columns=result.keys())
    display(df)
```

```python
with engine.begin() as conn:
    result = conn.execute(text("""
    SELECT title, COUNT(quantity)
    FROM books
    JOIN orders
    ON books.book_id = orders.book_id
    GROUP BY title;"""))
    df = pd.DataFrame(result.fetchall(), columns=result.keys())
    display(df)
```

```python
with engine.begin() as conn:
    result = conn.execute(text("""
    SELECT title, category_name
    FROM books
    INNER JOIN categories
    ON books.category_id = categories.category_id
    WHERE category_name = 'Biography';"""))
    df = pd.DataFrame(result.fetchall(), columns=result.keys())
    display(df)
```

```python
with engine.begin() as conn:
    result = conn.execute(text("""
    SELECT title, order_date
    FROM books
    JOIN orders
    ON books.book_id = orders.book_id
    WHERE order_date between '2024-06-01' AND '2024-06-02'"""))
    df = pd.DataFrame(result.fetchall(), columns=result.keys())
    display(df)
```

```python
df = pd.read_sql("""
    SELECT title, price
    FROM books
    ORDER BY price DESC""", engine)
display(df)
```

```python
with engine.begin() as conn:
    result = conn.execute(text("""
    SELECT title, AVG(price) as average_price
    FROM books
    GROUP BY title"""))
    df = pd.DataFrame(result.fetchall(), columns=result.keys())
    display(df)
```

```python
with engine.begin() as conn:
    result = conn.execute(text("""
    SELECT title, category
    FROM books
    JOIN categories
    WHERE (SELECT COUNT())"""))
    df = pd.DataFrame(result.fetchall(), columns=result.keys())
    display(df)
```
