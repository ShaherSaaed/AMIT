```python
class Flower:
  def __init__(self, name, color, height, is_blooming, water_level):
    self.name = name
    self.color = color
    self.height = height
    self.is_blooming = is_blooming
    self.water_level = water_level

  def grow(self):
    self.height += 5
  def water(self, amount):
    if (self.water_level + amount <= 100):
      self.water_level += amount
  def bloom(self):
    if (self.water_level >= 50):
      self.is_blooming = True
  def wilt(self):
    self.water_level -= 30
    self.is_blooming = False
  def display_info(self):
    print(f"Name: {self.name}")
    print(f"Color: {self.color}")
    print(f"height: {self.height}")
    print("Is blooming?", 'Yes' if self.is_blooming else 'No')
    print(f"Water level: {self.water_level}")
```

```python
metosa = Flower('Metosa', 'Purple', 32, False, 20)
metosa.display_info()
```

```python
metosa.water(50)
metosa.bloom()
metosa.display_info()
```

```python
class Flower:
    def __init__(self, color, size, **kwargs):
      super().__init__(**kwargs)
      self.color = color
      self.size = size

    def grow(self):
        print("The flower is growing.")

    def show_size(self):
        return self.size

class Fragrant:
    def __init__(self, smell_strength, **kwargs):
      super().__init__(**kwargs)
      self.smell_strength = smell_strength

    def describe_smell(self):
        print(f"The smell strength is {self.smell_strength}/10")


class Rose(Flower, Fragrant):
    def __init__(self, color, size, thorns, smell_strength):
      super().__init__(color=color, size=size, smell_strength=smell_strength)
      self.thorns = thorns

    def smell(self):
        print("The rose smells sweet.")

    def has_thorns(self):
        return self.thorns

r2 = Rose("Pink", 15, True, 8)
```

```python
class Animal:
  def __init__(self, name, age, species):
    self.name = name
    self.age = age
    self.species = species

  def speak(self):
    print("This animal makes a sound.")

  def show_info(self):
    print(f"Name: {self.name}")
    print(f"Age: {self.age}")
    print(f"Species: {self.species}")
```

```python
class Lion(Animal):
  def __init__(self, name, age, species, is_hungry):
    super().__init__(name, age, species)
    self.is_hungry = is_hungry
  def speak(self):
    print("Roar!")

  def feed(self):
    if(self.is_hungry):
      self.is_hungry = False
      print("The lion has been fed.")
    else:
      print("The lion is not hungry.")
```

```python
class Elephant(Animal):
  def __init__(self, name, age, species, trunk_length):
    super().__init__(name, age, species)
    self.trunk_length = trunk_length

  def speak(self):
    print("Trumpet!")

  def spray_water(self):
    print("The elephant sprays water with its trunk!")
```

```python
class Monkey(Animal):
  def __init__(self, name, age, species, is_playing):
    super().__init__(name, age, species)
    self.is_playing = is_playing
  def speak(self):
    print("Oh ooh aah aah!")

  def play(self):
    if (self.is_playing == False):
      self.is_playing = True
      print("The monkey starts playing")
    else:
      print("The monkey is already playing.")
```

```python
lion = Lion('Simpa', 12, 'Cats', True)
elephant = Elephant('Bob', 8,'Elephants', 3)
monkey = Monkey('Karl', 8, 'Monkeys', False)
```

```python
for animal in [lion, elephant, monkey]:
  animal.speak()
  animal.show_info()
  print('-----------------------')
```

```python
lion.feed()
elephant.spray_water()
monkey.play()
```

```python
class Person:

  counter = 0

  def __init__(self, name, age, national_id):
    self._name = name
    self._age = age
    self.__national_id = national_id
    Person.counter += 1

  def show_info(self):
    print(f"Name: {self._name}")
    print(f"Age: {self._age}")

  @property
  def national_id(self):
    return self.__national_id

  @national_id.setter
  def national_id(self, national_id):
    if(len(national_id) == 13):
      self.__national_id = national_id
    else:
      raise("Invalid national ID length!")
```

```python
class Doctor(Person):
  def __init__(self, name, age, national_id, specialty):
    super().__init__(name, age, national_id)
    self.specialty = specialty

  def show_info(self):
    super().show_info()
    print(f"Specialty: {self.specialty}")

  def diagnose(self):
    print(f" Dr. {self._name} is diagnosing patients...")
```

```python
class Patient(Person):
  def __init__(self, name, age, national_id, illness):
    super().__init__(name, age, national_id)
    self.illness = illness

  def show_info(self):
    super().show_info()
    print(f"Illness: {self.illness}")

  def take_medictation(self):
    print(f"{self._name} is taking medication for {self.illness}.")
```

```python
doctor1 = Doctor("House", 32, 1234567890123, 'Neurosurgery')
patient1 = Patient('Micheal', 28, 1234567890098, 'General weakness')
hospital = [doctor1, patient1]

for person in hospital:
  print(f"==={person._name}===")
  person.show_info()
  print(f"NationalID: {person.national_id}")
  if isinstance(person, Doctor):
    person.diagnose()
  elif isinstance(person, Patient):
    person.take_medictation()

print(f"Total Number of instants: {Person.counter}")

doctor1.national_id = '1234567891012'
doctor1.national_id
```
