```python
!pip install folium
```

```python
import folium
```

### Create a Simple Map

```python
city_map = folium.Map(location=[20,0], zoom_start=2)
```

```python
city_map
```

### Add City Markers

```python
# Loop through the first 100 cities in the dataset
for _, row in df.head(100).iterrows():
  # create a marker for each city
  folium.Marker(
      location=[row['lat'], row['lng']],
      tooltip=row['city']
  ).add_to(city_map)
```

```python
city_map
```

### Adjust Marker Size Based on Popilation

```python
for index, row in df.head(100).iterrows():
  folium.CircleMarker(
      location=[row['lat'], row['lng']],
      radius=min(row['population'] / 1_000_000, 10),
      popup=f"{row['city']} (Pop: {row['population']})"
  ).add_to(city_map)
```

```python
city_map
```

## Filter Cities by Country (USA) (Task)

```python
usa_map = folium.Map(location=[38.7946,-99.5348], zoom_start=4)
```

```python
usa_cities = df[df['iso3'] == 'USA'].head(200)
```

```python
for _, row in usa_cities.iterrows():
    folium.Marker(
        location=[row['lat'], row['lng']],
        tooltip=row['city']
    ).add_to(usa_map)
```

```python
for _, row in usa_cities.iterrows():
    folium.CircleMarker(
        location=[row['lat'], row['lng']],
        radius=min(row['population'] / 1_000_000, 10),
        popup=f"{row['city']}\n Population: {row['population']}"
    ).add_to(usa_map)
```
