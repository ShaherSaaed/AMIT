## Task 3

Advanced slicing using steps
 * Given a string `s = "abcdefghijklmn"`, print every second character.



````python
# TODO: Write your code here

s = "abcdefghijklmn"
# Your code
for i in range(0, len(s), 2):
  print(s[i], end="")
print("\nWith string methods: ")
print(s[::2])
````

## Task 4

Reverse string using slicing
 * Given `text = "Python"`, reverse it and print it.


````python
# TODO: Write your code here

text = "Python"
# Your code
for i in range(len(text) - 1, -1, -1):
  print(text[i], end='')
print("\nWith String methods:")
print(text[::-1])
````

**Description:**  
Build a simple password checker.

**Requirement:**  
Write a function that checks if a password meets the following rules:
- At least 8 characters  
- Contains both uppercase and lowercase letters  
- Contains at least one digit  


````python
# TODO: Write your code here

def check_password(password):
    # Your code here
    isAtLeastEight = False
    isUpper = False
    isLower = False
    isDigit = False
    if len(password) >= 8:
      isAtLeastEight = True
    for i in password:
      if(i.isupper()):
        isUpper = True
      if(i.islower()):
        isLower = True
      if(i.isdigit()):
        isDigit = True
    if isAtLeastEight and isUpper and isLower and isDigit:
      return True
    else:
      return False

# Example
print(check_password("P@ssword1"))  # Expected: True
print(check_password("12345"))
````

### ðŸŽ¯ Bonus Task : Palindrome Detector
**Description:**  
A palindrome is a word that reads the same backward.

**Requirement:**  
Write a function that checks if a word is a palindrome.


````python
# TODO: Write your code here

def is_palindrome(word):
    # Your code
    isPalindrome = True
    for i in range (len(word) // 2):
      if(word[i] != word[len(word) - i - 1]):
        isPalindrome = False
    return isPalindrome

print(is_palindrome("madam"))  # Expected: True
print(is_palindrome("edee"))
````
