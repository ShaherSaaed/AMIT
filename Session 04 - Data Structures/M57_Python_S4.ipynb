## Continue

```python
i = 0

while i <= 5:
  i = i + 1
  if i == 3:
    continue
  print(i)
```

4. Ask the user to enter a sentence.
Loop through each character and count how many vowels (a, e, i, o, u) it contains.



```python
text = input("Enter your sentence: ")

vowels = 0
for i in text:
  if i in ('a', 'e', 'i', 'o', 'u'):
    vowels += 1
print(vowels)
```

5. Write a function called is_prime(n) that returns True if n is a prime number, otherwise False.



```python
def is_prime(n):
  for i in range (2, n):
    if n % i == 0:
      return False
    else:
      return True
print(is_prime(9))
```

6. Write a function called factorial(n) that returns the factorial of n using a loop.

Example: factorial(5) → 120



```python
def factorial(num):
  fact = 1
  for i in range (num, 0, -1):
    fact *= i
  return fact

x = int(input("Enter a number: "))
print(factorial(x))
```

7. Given a list of words, use lambda and filter() to return a list of words with more than 5 letters.

Example: ["apple", "banana", "cherry", "fig", "grapes"] → ["banana", "cherry", "grapes"]

```python
lst = ["apple", "banana", "cherry", "fig", "grapes"]
filtered = list(filter(lambda  x :  len(x) > 5, lst))
print(filtered)
```

9. Write a function print_profile(name, **info) that:

Always prints the name

Then prints any other keyword information (age, city, job, etc.)

Example:`print_profile("Ali", age=23, city="Cairo", job="Developer")`


```python
def print_profile(name, **info):
  print(f"Name: {name}", end=", ")
  for key in info.keys():
    print(f"\"{key}\" = {info[key]}", end=", ")
print_profile("Ali", age=23, city="Cairo", job="Developer")
```
