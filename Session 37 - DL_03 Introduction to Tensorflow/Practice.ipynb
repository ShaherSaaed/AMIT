```python
import numpy as np
```

```python
v1 = tf.constant([1, 2, 3, 4])
v2 = tf.constant([6, 7, 8, 9])
```

```python
add_ = tf.add(v1, v2)
sub_ = tf.subtract(v1, v2)
multi_ = tf.multiply(v1, v2)
divide_ = tf.divide(v1, v2)
```

```python
for x in [add_, sub_, multi_, divide_]:
  with tf.compat.v1.Session() as sess:
    output = sess.run(x)
    print(output)
```
